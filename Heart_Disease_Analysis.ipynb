# Heart Disease Analysis

```python
# imports

%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# read datas from csv
heart_disease = pd.read_csv("data/heart-disease.csv")
heart_disease

```

```python
# plotting for over 50

over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
# using 00 method for plots
fig,ax= plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x = over_50["age"], y = over_50["chol"], c = over_50["target"]);
```

```python
#customize the plot
ax.set(title = "Heart Disease & Cholestrol Levels", Xlabel = "Age", ylabel = "Cholestrol");

# adding a legend
ax.legend(scatter.legend_elements(), title = 'Target');
```

```python
# imports

%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# plotting for over 50

over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()

# read datas from csv
heart_disease = pd.read_csv("data/heart-disease.csv")
heart_disease

#set the style
plt.style.use('seaborn-whitegrid')

# subplots of chol, age, thalch
fig,(ax0,ax1) = plt.subplots(nrows = 2, ncols = 1, figsize = (10,10), sharex = True )

# add data to ax0
scatter = ax0.scatter(x = over_50["age"], y = over_50["chol"], c = over_50["target"], cmap='winter')

# customize
ax0.set(title="Heart Disease & Cholestrol levels", Xlabel = "Age", ylabel = "Cholestrol")
ax0.set_xlim([50,80])

# add a legend
ax0.legend(*scatter.legend_elements(), title = "Target")

# add a meanline
ax0.axhline(y = over_50["chol"].mean(), linestyle = "--");

scatter = ax1.scatter(x = over_50["age"], y = over_50["thalach"], c = over_50["target"],cmap="plasma")

# customize
ax1.set(title="Heart Disease & Cholestrol levels", Xlabel = "Age", ylabel = "Heart-Rate")
ax1.set_ylim([60,200])


# add a legend
ax1.legend(*scatter.legend_elements(), title = "Target")

# add a meanline
ax1.axhline(y = over_50["thalach"].mean(), linestyle = "--");

# add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize = 16,fontweight = 'bold');



```
